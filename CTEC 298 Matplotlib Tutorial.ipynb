# Generating Basic Line Graph

```python
from matplotlib import pyplot as plt
```

```python
plt.plot([1, 2, 3], [4, 5, 1])
```

# Generating Line Graph with Title and Labels

```python
x = [5, 8, 10]
y = [12, 16, 6]

plt.title('Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.plot(x, y)
```

# Line Graph with Style

```python
from matplotlib import style
style.use('ggplot')
```

```python
x = [5, 8, 10]
y = [12, 16, 6]

x2 = [6, 9, 11]
y2 = [6, 15, 7]

plt.plot(x, y, 'g', label='line one', linewidth=5)
plt.plot(x2, y2, 'c', label='line two', linewidth=5)

plt.title('Epic Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.legend()

plt.grid(True, color='k')

plt.show()
```

# Creating a Bar Graph

```python
plt.bar([1, 3, 5, 7, 9], [5, 2, 7, 8, 2], label='Example one')
plt.bar([2, 4, 6, 8, 10], [8, 6, 2, 5, 6], label='Example two', color='g')

plt.legend()
plt.xlabel('bar number')
plt.ylabel('bar height')
plt.title('My plot yo!')

plt.show()
```

# Creating a Histogram

```python
population_ages = [
    22, 55, 62, 45, 21, 22, 34, 42, 42, 4, 99, 102, 110, 120, 121, 122, 130, 
    111, 115, 112, 80, 75, 65, 54, 44, 43, 42, 48
]

bins = [i*10 for i in range(13)]

plt.hist(population_ages, bins, histtype='bar', rwidth=0.8)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Histogram')
plt.legend()

plt.show()
```

# Making a Scatter Plot

```python
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [5, 2, 4, 2, 1, 4, 5, 2]

plt.scatter(x, y, label='skitcat', color='k', s=25, marker='o')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot')
plt.legend()

plt.show()
```

# Making a Stack Plot

```python
days = [1, 2, 3, 4, 5]

sleeping = [7, 8, 6, 11, 7]
eating = [2, 3, 4, 3, 2]
working = [7, 8, 7, 2, 2]
playing = [8, 5, 7, 8, 13]

plt.plot([], [], color='m', label='Sleeping', linewidth=5)
plt.plot([], [], color='c', label='Eating', linewidth=5)
plt.plot([], [], color='r', label='Working', linewidth=5)
plt.plot([], [], color='k', label='Playing', linewidth=5)

plt.stackplot(
    days, 
    sleeping, 
    eating, 
    working, 
    playing, 
    colors=['m', 'c', 'r', 'k']
)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Stack Plot')
plt.legend()

plt.savefig('stackPlot.png')
```

# Pie Chart

```python
slices = [7, 2, 2, 13]
activities = ['sleeping', 'eating', 'working', 'playing']
cols = ['c', 'm', 'r', 'b']

plt.pie(
    slices, 
    labels=activities, 
    colors=cols, 
    startangle=90, 
    shadow=True, 
    explode=(0, 0.1, 0, 0), 
    autopct='%1.1f%%'
)

plt.title('Pie Plot')

plt.show()
```

# Ploting Multiple Plots

```python
import numpy as np
```

```python
def f(t):
    return np.exp(-t) * np.cos(2 * np.pi * t)
```

```python
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2))

plt.subplot(212)
plt.plot(t2, np.cos(2 * np.pi * t2))

plt.show()
```
