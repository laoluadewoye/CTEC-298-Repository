```python
df8 = pd.read_csv(
    'stock_data.csv', 
    header=None, 
    names=['ticker', 'eps', 'revenue', 'price', 'people']
)
df8
```

```python
df9 = pd.read_csv('stock_data.csv', na_values=['not available', 'n.a.', -1])
df9
```

```python
df9.to_csv('new.csv', index=False)
```

```python
df9.to_csv('new_two_col.csv', columns=['tickers', 'eps'])
```

```python
def convert_people_cell(cell):
    if cell == "n.a.":
        return 'same walton'
    return cell

def convert_eps_cell(cell):
    if cell == "not available":
        return None
    return cell

df10 = pd.read_excel(
    'stock_data.xlsx', 
    converters = {
        'people': convert_people_cell,
        'eps': convert_eps_cell
    }
)
df10
```

```python
df10.to_excel(
    'new.xlsx', 
    sheet_name="stocks", 
    startrow=1, 
    startcol=2, 
    index=False
)
```

```python
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})

df_weather = pd.DataFrame({
    'day': ['1/1/2017', '1/2/2017', '1/3/2017'],
    'tempeerature': [32, 35, 28],
    'event': ['Rain', 'Sunny', 'Snow']
})

with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer, sheet_name="stocks")
    df_weather.to_excel(writer, sheet_name="weather")
```
